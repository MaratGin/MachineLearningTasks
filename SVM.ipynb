```python
import sys

import numpy as np
import numpy.linalg.linalg
import pygame
import random
import sklearn
from itertools import cycle
from math import hypot
from numpy import random
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

eps = 100

def compute_svm(data, target):
    model = SVC(kernel = 'linear')
    model.fit(data,target)
    return model
        

if __name__ == '__main__':
    pygame.init()
    screen = pygame.display.set_mode((800, 800), pygame.RESIZABLE)
    width = 800
    height = 800
    screen.fill(color='#FFFFFF')
    pygame.display.update()
    points = []
    test_points = []
    points_target = []
    test_target = []
    red_iter = 0
    blue_iter = 0
    X = 0
 
    
    flag = True
    drawing = False
    curColor = 0
    model = 0
    while flag:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                print(event)
                flag = False
            if event.type == pygame.MOUSEBUTTONDOWN and event.button == 1:
                drawing = True
                curColor = 0
                print("pressed left mousebutton")
            if event.type == pygame.MOUSEBUTTONDOWN and event.button == 3:
                drawing = True
                curColor = 1
                print("pressed right mousebutton")
            pygame.display.update()
            # if event.type == pygame.KEYDOWN:
            #     if event.key == 13:
                    # screen.fill(color='#000000')
            if event.type == pygame.MOUSEBUTTONUP and event.button == 1:
                drawing = False
                print("unpressed left mousebutton")
            if event.type == pygame.KEYDOWN and event.key == pygame.K_2:
                print("New point task")
                coord = pygame.mouse.get_pos()
                newArr = [[coord[0],coord[1]]]
                newValue = model.predict(np.array(newArr))
                if newValue == 0:
                    pygame.draw.circle(screen, color='red', center=pygame.mouse.get_pos(), radius=10)
                if newValue == 1:
                    pygame.draw.circle(screen, color='blue', center=pygame.mouse.get_pos(), radius=10)
                    
                
            if event.type == pygame.MOUSEBUTTONUP and event.button == 3:
                drawing = False
                print("unpressed right mousebutton")
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_1:
#                     clusters = dbscan_naive(points, eps, 3)
                    print('----------')
                    X = np.array(points)
                    print(X)
                    print(points_target)
                    model = compute_svm(X,points_target)
                    w = model.coef_[0]
                    b = model.intercept_[0]

                    xx = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
                    x_points = np.linspace(0, 1000)
                    y_points = -(w[0] / w[1]) * x_points - b / w[1]
                    pygame.draw.aaline(screen, 'green', [x_points[0], y_points[0]], [x_points[len(x_points) - 1], y_points[len(y_points) - 1]])

            pygame.display.update()
        if drawing:
            pygame.time.delay(60)
            coord = pygame.mouse.get_pos()
            print(coord[0], coord[1])
            if curColor == 0:
                pygame.draw.circle(screen, color='red', center=pygame.mouse.get_pos(), radius=10)
                print(len(points) % 5)
                if red_iter % 5 == 0:
                    print('test added')
                    test_points.append([coord[0],coord[1]])
                    test_target.append(1)
                    red_iter = red_iter + 1
                else:
                    print('main added')
                    points.append([coord[0],coord[1]])
                    points_target.append(0)
                    red_iter = red_iter + 1

            if curColor == 1:
                pygame.draw.circle(screen, color='blue', center=pygame.mouse.get_pos(), radius=10)
                if blue_iter % 5 == 0:
                    test_points.append([coord[0],coord[1]])
                    test_target.append(1)
                    blue_iter = blue_iter + 1
                else:
                    points.append([coord[0],coord[1]])
                    points_target.append(1)
                    blue_iter = blue_iter + 1

            
            # print("points - ", points)
    pygame.quit()
    sys.exit()
```
